```python
from typing import List

nums = [5, 8, 23, 7, 1]
target = 15
def twoSum(self, nums: List[int], target: int) -> List[int]:
        hashtable = dict()
        for i, num in enumerate(nums):
            if target - num in hashtable:
                return [hashtable[target - num], i]
            hashtable[nums[i]] = i
        return []

result = twoSum(None, nums, target)
print(result)
```

```python
from collections import defaultdict


nums = [2,2,1,1,1,2,2]
dic = defaultdict(int)
for num in nums:
    dic[num] += 1

# return出现大于 len(nums) / 2 次的元素
for num, count in dic.items():
    if count > len(nums) / 2:
        print(num)
        break
```

```python
nums = [7,1,5,3,6,4]
# find the lowest number's index
min_index = 0
for i in range(1, len(nums)):
    if nums[i] < nums[min_index]:
        min_index = i
print(min_index)
```
