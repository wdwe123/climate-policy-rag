```python
import json
import random
from collections import defaultdict

jsonl_path = "E:/2026_capstone/policy_data/output/final_chunks.jsonl"

with open(jsonl_path, "r", encoding="utf-8") as f:
    chunks = [json.loads(line) for line in f if line.strip()]

print(f"Total chunks: {len(chunks)}")


problem_buckets = {
    "no_heading": [],
    "very_small": [],
    "very_large": [],
    "appendix": [],
    "short_section_title": [],
}

for ch in chunks:
    tokens = ch.get("tokens", 0)
    section = ch.get("section", "")

    if section == "(no heading)":
        problem_buckets["no_heading"].append(ch)

    if tokens < 200:
        problem_buckets["very_small"].append(ch)

    if tokens > 1400:
        problem_buckets["very_large"].append(ch)

    if ch.get("is_appendix", False):
        problem_buckets["appendix"].append(ch)

    if section and len(section) < 8:
        problem_buckets["short_section_title"].append(ch)


def print_chunk(ch, max_chars=800):
    print("---")
    print(f"doc_title: {ch['doc_title']}")
    print(f"section: {ch['section']}")
    print(f"pages: {ch['pages']}")
    print(f"is_appendix: {ch['is_appendix']} appendix_label: {ch['appendix_label']}")
    print(f"tokens: {ch['tokens']}")
    print(ch["text"][:max_chars])
    if len(ch["text"]) > max_chars:
        print("...")

for bucket, bucket_chunks in problem_buckets.items():
    if not bucket_chunks:
        continue

    print(f"\n================ {bucket.upper()} ({len(bucket_chunks)}) ================\n")
    sample = random.sample(bucket_chunks, k=min(3, len(bucket_chunks)))
    for ch in sample:
        print_chunk(ch)
        print()
```

```python
import json
from collections import defaultdict

jsonl_path = "E:/2026_capstone/policy_data/output/final_chunks.jsonl"

with open(jsonl_path, "r", encoding="utf-8") as f:
    chunks = [json.loads(line) for line in f if line.strip()]

by_doc = defaultdict(list)
for ch in chunks:
    by_doc[ch["doc_title"]].append(ch)


for doc_title, doc_chunks in by_doc.items():
    print(f"\n================ FILE: {doc_title} ================\n")

    for ch in doc_chunks[:10]:
        print("---")
        print(f"section: {ch['section']} | pages: {ch['pages']} | tokens: {ch['tokens']}")
        print(ch["text"][:600])
        print("...\n")
    break
```
